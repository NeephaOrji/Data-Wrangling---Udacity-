```python
#import packages
import pandas as pd
import numpy as np
from matplotlib  import pyplot as plt
% matplotlib inline
```

### Question 1: Which dog stage has the most likes/ favorite count

```python
# Read the twitter file as df
df = pd.read_csv('twitter_archive_master.csv')
```

```python
# Create colors for the visualization
colors = ['Grey', 'Grey', 'Grey', 'Grey', 'Grey', 'Grey', 'Grey', 'Blue']
```

```python
# Use the dog stages column and the favorite column to answer the above stated question
df.groupby('dog_stages')['favorite_count'].mean().sort_values(ascending = True).plot(kind = 'barh', color = colors, figsize=(15,10))
plt.title('Average Likes/ favorite count by Dog Stages')
plt.xlabel('Likes')
plt.ylabel('Dog stages');
```

From the output above, we can tell that the dog stage with the most likes or favorite count is puppo-doggo, followed by puppo, doggo, floffer_doggo, floffer, doggo-pupper, none and pupper. So the dog stage with the least likes is pupper. Onto the next insight.

### Question 2: Which dog stage has the most retweets / retweet count

```python
# Create colors for the visualization
colors = ['Grey', 'Grey', 'Grey', 'Grey', 'Grey', 'Grey', 'Grey', 'Blue']
```

```python
# Use the dog stages column and the retweet count column to answer the above stated question
df.groupby('dog_stages')['retweet_count'].mean().sort_values(ascending = True).plot(kind = 'barh', color = colors, figsize=(15,10))
plt.title('Average retweet/ retweet count by Dog Stages')
plt.xlabel('Retweets / Retweet_count')
plt.ylabel('Dog stages');
```

From the output above, we can tell that the dog stage with the most retweets or retweet count is puppo-doggo, followed by doggo, puppo, floffer, doggo-pupper, floffer_doggo, none and pupper. So the dog stage with the least likes is pupper. This begs the question if there is a correlation between likes and retweets or favorite count and retweet counts. Is there?

### Question 3: Is there a Correlation between Likes and Retweets or Likes and Retweet Count

```python
# Use the retweet count and favourite count column to create a scatter plot
df.plot(x = 'retweet_count', y = 'favorite_count', kind = 'Scatter', figsize =(15, 10))
plt.title('Correlation between Likes and Retweets')
plt.xlabel('Retweets')
plt.ylabel('Likes');
```

Did you notice that as retweets increases, likes increases too?, this suggest that there is a positive relationship between retweets/ retweet count and likes/favourite count.

### Question 4:  How is the source tweet shared and which mode has the highest percentage?

```python
# Use the source column to make a pie chart
df['source'].value_counts().plot(kind = 'pie', figsize =(10,8))
plt.title('Tweet Source');
```

The visualization above shows that most tweet were made from iphones followed by Twitter Web Client and Tweet Deck having the least percentage.
